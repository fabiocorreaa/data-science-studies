```python
import seaborn as srn
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('co2.csv')
base.head()
```

```python
srn.scatterplot(base.conc, base.uptake, hue=base.Type)
```

```python
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(q.conc, q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
srn.scatterplot(m.conc, m.uptake).set_title('Mississipi')
plt.tight_layout()
```

```python
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(q.conc, q.uptake).set_title('Chilled')
plt.subplot(1,2,2)
srn.scatterplot(m.conc, m.uptake).set_title('Non chilled')
plt.tight_layout()
```
